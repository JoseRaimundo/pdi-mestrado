## Peak Signal-to-Noise Ratio


A Peak Signal-to-Noise Ratio (PSNR), é uma das métricas mais utilizadas na literarura, apesar de ter uma abordagem simples, ela retorna informações importantes sobre saturações que normalmente são imperceptíveis para o olho humano. Ela avalia a relação entre a potência máxima possível de uma imagem e a potência de ruído corruptor que afeta a qualidade de sua representação (ou imagem nova). 

PSNR é definido da seguinte forma:


![psnr](tamplete/psnr.svg)


Em que L é o número de níveis de intensidade máximos possíveis em uma imagem. E o Mean Square Error (MSE) é o erro quadrático médio que é definido como:

![mse](tamplete/mse.svg)

```python
## Bibliotecas para manipulação de dados e matemática
from math import log10, sqrt 

### Biblioteca para visualização
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
```

```python
# Carrega um arquivo de imagem em uma matriz de formato mxnx3 (altura x largura x canais de cores RGB)
original = mpimg.imread('img/mario.jpg')

# Informações da image
print("Tamanho: ", original.shape[0] ,' x ', original.shape[1]  )
plt.imshow(original)
plt.show()

img_nova = mpimg.imread('img/mario_ruido.jpg')

print("Tamanho: ", img_nova.shape[0] ,' x ', img_nova.shape[1]  )
plt.imshow(img_nova)
plt.show()
```

```python
def PSNR(original, nova): 
    mse = np.mean((original - nova) ** 2) 
    if(mse == 0):  # Se o MSE for 0 (zero), então não há ruído na imagem, PSNR = 100
        return 100
    max_pixel = 255.0
    psnr = 20 * log10(max_pixel / sqrt(mse)) 
    return psnr
```

```python
print(f'PSRN da imagem original com ela mesma: {PSNR(original, original)}')
```

```python
print(f'PSRN da imagem original com a modificada: {PSNR(original, img_nova)}')
```
