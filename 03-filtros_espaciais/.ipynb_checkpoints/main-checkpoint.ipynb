```python
## Função para converter RGB para tons de cinza 
def rgb2gray(rgb):
    return np.dot(rgb[...,:3], [0.299, 0.587, 0.144])
```

```python
def filtroMedia(original):   
    altura_original  = img_gray.shape[0]
    largura_original = img_gray.shape[1]
    
    # Kernel 2 x 2 (matriz de 2x2 preenchida com zeros)
    janela = np.zeros((3,3), dtype=np.float64)
    
    # Matriz para colocar os valores dos pixels resultante do filtro
    destino = np.zeros((round(altura_original/2),round(largura_original/2)), dtype=np.float64)

    ## Cordenadas do pixels de destino vai receber o resultado do filtro
    destino_linha = 0
    destino_coluna = 0
    
    # Percorre a imagem original: linhas
    ## Obs.: Note que o 'for' vai de zero até a altra da imagem original e 
    ## Com passo 2, ou seja, vai icrementar 2, 4, 6, 8... Isso é o janelamento
    ## Ou seja, a cada 2 pixels eu começo a pegar os pixels de uma janela nova
    for original_linha in range(0, altura_original-2, 2):
        destino_coluna = 0
        # Percorre a imagem original: colunas
        for original_coluna in range(0, largura_original-2, 2):
            janela[0][0] = img_gray[original_linha][original_coluna]
            janela[0][1] = img_gray[original_linha][original_coluna+1]
            janela[0][2] = img_gray[original_linha][original_coluna+2]
            
            janela[1][0] = img_gray[original_linha+1][original_coluna]
            janela[1][1] = img_gray[original_linha+1][original_coluna+1]
            janela[1][2] = img_gray[original_linha+1][original_coluna+2]
            
            janela[2][0] = img_gray[original_linha+2][original_coluna]
            janela[2][1] = img_gray[original_linha+2][original_coluna+1]
            janela[2][2] = img_gray[original_linha+2][original_coluna+2]
            destino[destino_linha][destino_coluna] = janela.mean()
            
            destino_coluna = destino_coluna + 1
        destino_linha = destino_linha + 1
    
    
    print( f"Tamanho da imagem Original: {img_gray.shape[0]} x {img_gray.shape[1]}")
    plt.imshow(img_gray, cmap = 'gray')
    plt.show()      
    
    print( f"Tamanho da imagem final: {destino.shape[0]} x {destino.shape[1]}")
    print( f"Tamanho da imagem final: {destino_linha} x {destino_coluna}")

    plt.imshow(destino, cmap = 'gray')
    plt.show()        
```
